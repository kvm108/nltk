```python
from nltk.corpus import movie_reviews
from random import shuffle
from nltk import FreqDist
from nltk import NaiveBayesClassifier
from nltk import classify
```

<h3> 0. Dump data </h3>

```python
print('Category')
for category in movie_reviews.categories():
    print(category)
    
# print("File Id's")
# for fileid in movie_reviews.fileids():
#     print(fileid, movie_reviews.words(fileid))
    
```

<h3> 1. Get data and tokenize words </h3>

```python
#document - tokenize words, class

document = [ (movie_reviews.words(fileid), category)
    for category in movie_reviews.categories()
    for fileid in movie_reviews.fileids(category) ]
    
shuffle(document)
```

<h3> 2. Features </h3>

- One features for every word? Overkill
- Let's choose Top 2000 Frequent words
- Make the feature extractor for feature set

```python
all_words = FreqDist(w.lower() for w in movie_reviews.words())
len(all_words)
```

^ that's too many features to have

```python
word_features1 = list(all_words)[:2000]

def document_fetaures(document):
    doc_words = set(document) # comapring *in* in set faster than list 
    features = {}
    
    for word in word_features1:
        features['contains({})'.format(word)] = (word in doc_words)
    return features

document_fetaures(movie_reviews.words('pos/cv957_8737.txt'))
```

```python
top_two_thousand = all_words.most_common(2000)

def document_fetaures_top(document):
    doc_words = set(document) # comapring *in* in set faster than list 
    features = {}
    
    for word in top_two_thousand:
        features['contains({})'.format(word)] = (word in doc_words)
    return features

document_fetaures_top(movie_reviews.words('pos/cv957_8737.txt'))
```

<h3> 3. Generate feature set using document_feature </h3>

```python
featuresets = [(document_fetaures(doc), category) for (doc, category) in document]
featuresets[:1]

featuresets_top = [(document_fetaures_top(doc), category) for (doc, category) in document]
featuresets_top[:1]

train, test = featuresets[100:], featuresets[:100]
train_top, test_top = featuresets_top[100:], featuresets_top[:100]
```

<h3> 4. Train test </h3>

```python
clf = NaiveBayesClassifier.train(train)
print("first 2000 ", classify.accuracy(clf, test))

clf_top = NaiveBayesClassifier.train(featuresets_top)
print("most common 2000 ",classify.accuracy(clf_top, test_top))
```
